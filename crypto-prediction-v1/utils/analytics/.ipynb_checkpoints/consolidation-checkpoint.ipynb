```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mplcursors
import mpld3
```

```python
df = pd.read_csv("../../datasets/binance-bitcoin.csv")
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Close Time'] = pd.to_datetime(df['Close Time'], format='mixed').dt.strftime('%Y-%m-%d %H:%M:%S')
df['Close Time'] = pd.to_datetime(df['Close Time'], format='%Y-%m-%d %H:%M:%S')
```

```python
print(df[-100:-1])
```

```python
# Предполагая, что у вас есть DataFrame с названием df и столбцами 'Date' и 'Close'
# Сначала преобразуйте столбец 'Date' в тип данных datetime
df['Date'] = pd.to_datetime(df['Date'])

# Отфильтруйте данные, чтобы получить только нужный временной интервал
start_date = pd.to_datetime('2023-10-25')
end_date = pd.to_datetime('2023-11-08')
filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

# Постройте график только для отфильтрованных данных
plt.plot(filtered_df['Date'], filtered_df['Close'])
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Stock Price')

# Создайте всплывающую подсказку с информацией о точной цене
tooltip = mpld3.plugins.PointLabelTooltip(plt.gca(), labels=np.round(filtered_df['Close'], 2).astype(str).tolist())
mpld3.plugins.connect(plt.gcf(), tooltip)

# Преобразуйте график в "виджет" для интерактивной визуализации
html_graph = mpld3.fig_to_html(plt.gcf())

plt.close()  # Закрываем график, так как нам нужен только HTML-код

# Теперь вы можете использовать `html_graph` для вставки интерактивного графика на вашу веб-страницу

# Пример использования в Jupyter Notebook:
from IPython.display import HTML
HTML(html_graph)

# Пример использования в HTML-файле:
# with open('graph.html', 'w') as f:
#    f.write(html_graph)
```
